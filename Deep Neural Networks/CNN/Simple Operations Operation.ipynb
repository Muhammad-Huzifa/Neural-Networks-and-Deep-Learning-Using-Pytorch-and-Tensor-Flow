**Drivatives**

```python
X= torch.tensor(2.0,requires_grad=True)
X
```

```python
y = X**2
y.backward()
X.grad
```

```python
x = torch.tensor(2.0,requires_grad=True)
z = x**2 + 2*x + 1
z.backward()
x.grad
```

```python
u = torch.tensor(1.0,requires_grad=True)
v = torch.tensor(2.0,requires_grad=True)
f = u*v + u**2
f.backward()
u.grad
v.grad
```
